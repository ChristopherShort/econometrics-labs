Task 7: Regression and correlation:
-----------------------------------

Let's look at the relationship between the log of GDP per worker in 1960 and growth in the subsequent 25 years...

```python
fig = plt.figure()
ax = fig.add_subplot(111)

X = mrw_1992_data_nonoil.lngdp60.values
Y = mrw_1992_data_nonoil.growth.values

ax.scatter(X, Y, color='orange', marker='o')

ax.set_xlabel('log real GDP per worker (1960)')
ax.set_ylabel('Average growth in GDP 1960-1985')
ax.set_title('How fast did you grow?')

plt.show()
```

It's not so obvious (to me anyway) how strong the relationship is between these two variables (or even what it might have!), so let's run a regression to find out!

We will be using the Python [statsmodels](http://statsmodels.sourceforge.net/devel/index.html) package to estimate nearly all of our models throughout the year.  The command index for stats models can be found [here](http://statsmodels.sourceforge.net/devel/genindex.html).  For now focus on the [OLS](http://statsmodels.sourceforge.net/devel/generated/statsmodels.regression.linear_model.OLS.html#statsmodels.regression.linear_model.OLS) command. More information on regression models can be found [here](http://statsmodels.sourceforge.net/devel/regression.html).

```python
# our endogenous (or dependent or response variable) is growth
y = mrw_1992_data_nonoil['growth']

# our matrix of exogenous (or independent variables) are a constant and lngdp60
X = sm.add_constant(mrw_1992_data_nonoil['lngdp60'], prepend=True)

# run the regression
results = sm.OLS(endog=y, exog=X).fit()

# examine the output
print results.summary()
```

OK, so the coefficient we get is 0.0038 with a p-value of 0.06. What does this mean? In purely statistical terms, it simply means that the value of the slope coefficient is positive and marginally statistically significant. What does this analysis imply about convergence of GDP per worker across countries? The fact that the slope coefficient is positive implies that GDP per worker is diverging rather than converging! It looks like countries with high GDP per capita in 1960 went on to have higher average growth rates as well!

Let's look at the scatterplot again, but this time overlay the best linear fit (as determined by OLS)...

```python
fig = plt.figure()
ax = fig.add_subplot(111)

X = mrw_1992_data_nonoil.lngdp60.values
Y = mrw_1992_data_nonoil.growth.values

# the scatterplot
ax.scatter(X, Y, color='orange', marker='o')

# doesn't seem to be a command to plot a line using slope and intercept!
grid = np.linspace(5, 10.5, 1000)
ax.plot(grid, results.params[1] * grid + results.params[0], color='red', label='OLS')

ax.set_xlim(5.5, 10.5)
ax.set_xlabel('log real GDP per worker (1960)')
ax.set_ylabel('Average growth in GDP 1960-1985')
ax.set_title('Divergence of GDP per worker?')

ax.legend(loc='best', frameon=False)

plt.show()
```

Dpes this mean that poor countries are simply doomed? No! Before we can conclude anything about the relationship between growth and the initial level of GDP we need to control for other relevant variables.

Let's look again at our primary variables of interest.  Another. non-graphical way to look at relationships between variables is by examining a correlation matrix.

```python
# Create a correlation matrix
mrw_1992_data_nonoil[['growth', 'lngdp60', 'lngdp85', 'popgrowth', 'igdp', 'school']].corr()
```

From the above correlation matrix you can see that growth isn't just correlated with the level of GDP per worker in 1960! GDP is also positively correlated with rates of investment and enrollment in secondary eduction, and is negatively correlated with population growth rates.  Given the strength of these correlations, it is entirely possible that the relationship we thought we saw between growth and initial GDP was actually the result of these other relationships that we totally ignored!

What happens if we include these other variables in the regression...

```python
# our endogenous (or dependent or response variable) is growth
y = mrw_1992_data_nonoil['growth']

# our matrix of exogenous (or independent variables) are a constant and lngdp60
X = sm.add_constant(mrw_1992_data_nonoil[['lngdp60', 'popgrowth', 'igdp', 'school']], prepend=True)

# run the regression
results = sm.OLS(endog=y, exog=X).fit()

# examine the output
print results.summary()
```

Look at that! The relationship between growth and initial GDP per worker has completely reversed: it used to be positive and marginally significant, and it's now negative and highly significant!  This means that the positive relationship we observed between growth and initial level of GDP in the bivariate regression was all wrong. It just looked positive because the poor countries were initially also the ones with low rates of investment and school enrollment.  These countries grew slowly in subsequent years because they underinvested in physical and human capital! The multiple regression analysis implies that there is convergence, but only conditional convergence of GDP per worker.
