```python
import numpy as np
from scipy import stats
import pandas as pd
import statsmodels.api as sm
import matplotlib as mpl
import matplotlib.pyplot as plt
```

###Part 1:

Some notes on this DGP.  Note that expected value of the disturbance term is non-zero:

$$ E[u] = E[e] + 0.3E[x] - 0.15 = -0.15 + 0.3E[x] \neq 0$$

In addition, the disturbance term is correlated with the regressor:

$$ E[x^{T}u] = E[x^{T}(e + 0.3(x - 0.5))] = E[x^{T}e] + 0.3E[x^{T}x] - 0.5E[x^{T}] = 0.3E[x^{T}x] - 0.5E[x^{T}] \neq 0$$

Easiest way to see that there is no heteroskedasticity is to note that the DGP can be re-written as follows

$$ y = 1 + x + u = 1 + x + (e + 0.3(x - 0.5)) = 0.85 + 1.3x + e$$

which is clearly homoskedastic.  When we regress $y$ on $x$ we are the OLS estimator is going to consistently (and efficiently estimate) the slope to be 1.3 instead of 1! In this case, because the regressor is correlated with the error term, the OLS estimator will be biased (but because the error is homoskedastic OLS will be efficiently biased!).

```python
# Create a random variable object representing the normal distribution
mu = 0
sigma = 1
normal_rv = stats.distributions.norm(mu, sigma)

# Create a random variable object representing a uniform distribution
uniform_rv = stats.distributions.uniform()
```

```python
#### First, create the fake data and run the regression #####

# set the number of obs
N = 500

# create and error term (note that I convert an ndarray to a Series object!)
e = pd.Series(normal_rv.rvs(N), name='e')

# create an independent variable
x = pd.Series(uniform_rv.rvs(N), name='x')

# create a disturbance term
u = pd.Series(e + 0.3 * (x - 0.5), name='u')

# create a dependent variable y
y = pd.Series(1 + x + u, name='y')

# prepend a constant to the independent variable to create a matrix of regressors
X = sm.add_constant(x, prepend=True)

# create the model object
model = sm.OLS(endog=y, exog=X)

# fit the model and save the the results
results = model.fit()

# print the results
print results.summary()
```

Note the extremely large $t$-stats for both the constant and the regressor.  This should be a tip-off that something is not right.

```python
fig = plt.figure()

# first subplot is a histogram of x
fig.add_subplot(131)
X.x.hist(color='r', normed=True, alpha=0.75, bins=15)
plt.title('Uniform!')
plt.xlabel('x')
plt.ylabel('Density')

# second subplot is a histogram of the disturbance
fig.add_subplot(132)
u.hist(color='g', normed=True, alpha=0.75, bins=15)
plt.title('??')
plt.xlabel('u')

# final subplot is a histogram of y
fig.add_subplot(133)
y.hist(color='b', normed=True, alpha=0.75, bins=15)
plt.title('??')
plt.xlabel('y')

fig.tight_layout()

plt.show()
```

Calculate the correlation between our regressor and the disturbance term, and then construct a test of the null hypothesis that the regressor and the distrubance term are uncorrelated.

```python
# how correlated are the regressor and the distrubance
x.corr(u)
```

```python
# t-stat for the null hypothesis that x and u are uncorrelated...
t_stat = x.corr(u) / np.sqrt((1 - x.corr(u)**2) / (N - 2))
print t_stat
```

```python
# critical value for a 5% test with N-2 degrees of freedom
stats.distributions.t.ppf(1 - (0.05 / 2), N - 2)
```

I would reject the null hypothesis that our regressor and disturbance term are uncorrelated! At the 5% level of significance.  Now lets scatter plot of the data and see if we can spot the heteroskedasticity...

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# create a scatterplot
ax.plot(X.x.values, y.values, 'bo', alpha=0.25)
ax.set_xlabel('Dependent variable, x')
ax.set_ylabel('Independent variable, y')
ax.set_title('Homoskedasticity?', fontsize=20)

# add the regression line
grid = np.linspace(0, 1, 1000)
ax.plot(grid, results.params['const'] + results.params['x'] * grid, 'r-', label='OLS')
ax.legend(loc='best', frameon=False)

plt.show()
```

```python
# set the number of obs in each sample
N = 500 

# set the number of replications (best to start with a small number for debugging purposes)
reps = 10000

# create arrays to store estimates of slope and its SE
beta_hats = pd.Series(np.zeros(reps), name='beta')

# for loop conducts the MC
for rep in range(reps):
    # Gaussian noise term with mu=0, sigma = 1
    e = pd.Series(stats.distributions.norm(0, 1).rvs(N), name='error')
    # Uniform independent variable on [0, 1]
    x = pd.Series(stats.distributions.uniform(0, 1).rvs(N), name='x')
    # disturbance term u. Note that moments of disturbance term depend on x!
    u = pd.Series(e + 0.3 * (x - 0.5), name='disturbance')
    
    # Create dependent variable y
    y = pd.Series(1 + x + u, name='y')
    
    # regress y on x
    X = sm.add_constant(x, prepend=True)
    model = sm.OLS(endog=y, exog=X)
    results = model.fit()
    
    # extract estimate of slope
    beta_hat = results.params['x']
    
    # store the estimates
    beta_hats[rep] = beta_hat
    
    # conditional statement tells us how many iterations have completed
    if (rep + 1)% 500 == 0:
        print "Done with %i out of %i iterations." % (rep + 1, reps)
        
```

Plotting the histogram of our estimates for $\hat{\beta}$, we see that the OLS estimator is biased as expected.  In this case we should also have expected the OLS estimator to be biased upwards.  Why? The variable $x$ is positively correlated with both $y$ and $u$. 

```python
# create new figure and axes objects
fig = plt.figure()
ax = fig.add_subplot(111)

# plot the histogram
n, bins, patches = ax.hist(beta_hats, bins=40, normed=True, color='blue', alpha=0.75)
ax.set_xlabel(r'$\hat{\beta}$', fontsize=15)
ax.set_ylabel('Density', family='serif', fontsize=15)
ax.set_title('%i replications of N=%i observations' %(reps, N), family='serif', fontsize=15)

# overlay the normal density
mu = np.mean(beta_hats)
sigma = np.std(beta_hats)
grid = np.linspace(np.min(beta_hats), np.max(beta_hats), 1000)
ax.plot(grid, stats.norm(mu, sigma).pdf(grid), 'r-', label=r'$\mathcal{N}$')
ax.set_ylim(0, 3)

# demarcate the true value of beta
ax.vlines(x=1, ymin=0, ymax=3, linestyle='dashed', label=r'$\beta$')

# add a legend
ax.legend(loc='best', frameon=False)

# save the plot
plt.savefig('Talk about a biased estimator.png')

# display plot
plt.show()
```

The moral of this story? Correlation between error term and the regressor causes OLS to be biased!

###Part 2:

Some notes on this DGP.  The problem is the opposite as in part 1.  The expected value of the disturbance term is zero...

$$ E[ex] = E[e]E[x] - Cov(e, x) = 0$$

...and the disturbance term is uncorrelated with the regressor. Thus the OLS estimator will be unbiased and consistent:

$$ E[x^Tu] = E[x^T(ex)] = E[e]E[x^{T}x] - Cov(e, x^{T}x) = 0 $$

However, the variance of the disturbance term depends on the value of the regressor which implies heteroskedasticity (a violation of classic OLS).

```python
# set the number of obs
N = 500

# create and error term (note that I simply change the args passed to normal_rv!)
normal_rv.args=(0, 4)
e = pd.Series(normal_rv.rvs(N), name='e')

# create an independent variable
x = pd.Series(uniform_rv.rvs(N), name='x')

# create a disturbance term
u = pd.Series(e * x, name='u')

# create a dependent variable y
y = pd.Series(1 + x + u, name='y')

# prepend a constant to the independent variable to create a matrix of regressors
X = sm.add_constant(x, prepend=True)

# create the model object
model = sm.OLS(endog=y, exog=X)

# fit the model and save the the results
results = model.fit()

# print the results
print results.summary()
```

```python
fig = plt.figure()

# first subplot is a histogram of x
fig.add_subplot(131)
X.x.hist(color='r', normed=True, alpha=0.75, bins=15)
plt.title('Uniform!')
plt.xlabel('x')
plt.ylabel('Density')

# second subplot is a histogram of the disturbance
fig.add_subplot(132)
u.hist(color='g', normed=True, alpha=0.75, bins=15)
plt.title('non-normal!')
plt.xlabel('u')

# final subplot is a histogram of y
fig.add_subplot(133)
y.hist(color='b', normed=True, alpha=0.75, bins=15)
plt.title('non-normal!')
plt.xlabel('y')

fig.tight_layout()

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# create a scatterplot
ax.plot(X.x.values, y.values, 'bo', alpha=0.25)
ax.set_xlabel('Dependent variable, x')
ax.set_ylabel('Independent variable, y')
ax.set_title('Obvious heteroskedasticity!', fontsize=20)

# add the regression line
grid = np.linspace(0, 1, 1000)
ax.plot(grid, results.params['const'] + results.params['x'] * grid, 'r-', label='OLS')
ax.legend(loc='best', frameon=False)

plt.show()
```

```python
# set the number of obs in each sample
N = 500 

# set the number of replications (best to start with a small number for debugging purposes)
reps = 10000

# create arrays to store estimates of slope and its SE
beta_hats = pd.Series(np.zeros(reps), name='beta')
se_beta_hats = pd.Series(np.zeros(reps), name='se')
t_stats = pd.Series(np.zeros(reps), name='t-stat')

# for loop conducts the MC
for rep in range(reps):
    # Gaussian noise term with mu=0, sigma = 1
    e = pd.Series(stats.distributions.norm(0, 4).rvs(N), name='error')
    # Uniform independent variable on [0, 1]
    x = pd.Series(stats.distributions.uniform(0, 1).rvs(N), name='x')
    # disturbance term u. Note that moments of disturbance term depend on x!
    u = pd.Series(e * x, name='disturbance')
    
    # Create dependent variable y
    y = pd.Series(1 + x + u, name='y')
    
    # regress y on x
    X = sm.add_constant(x, prepend=True)
    model = sm.OLS(endog=y, exog=X)
    results = model.fit()
    
    # extract estimate of slope
    beta_hat = results.params['x']
    # extract the se for the estimate
    se_beta_hat = results.bse['x']
    # extract the test statistic
    t_stat = (beta_hat - 1) / se_beta_hat
    
    # store the estimates
    beta_hats[rep] = beta_hat
    se_beta_hats[rep] = se_beta_hat
    t_stats[rep] = t_stat
    
    # conditional statement tells us how many iterations have completed
    if (rep + 1)% 500 == 0:
        print "Done with %i out of %i iterations." % (rep + 1, reps)
        
```

```python
# create new figure and axes objects
fig = plt.figure()
ax = fig.add_subplot(111)

# plot the histogram
n, bins, patches = ax.hist(beta_hats, bins=40, normed=True, color='blue', alpha=0.75)
ax.set_xlabel(r'$\hat{\beta}$', fontsize=15)
ax.set_ylabel('Density', family='serif', fontsize=15)
ax.set_title('%i replications of N=%i observations' %(reps, N), family='serif', fontsize=15)

# overlay the normal density
mu = np.mean(beta_hats)
sigma = np.std(beta_hats)
grid = np.linspace(np.min(beta_hats), np.max(beta_hats), 1000)
ax.plot(grid, stats.norm(mu, sigma).pdf(grid), 'r-', label=r'$\mathcal{N}$')
ax.set_ylim(0, 1.5)

# demarcate the true value of beta
ax.vlines(x=1, ymin=0, ymax=3, linestyle='dashed', label=r'$\beta$')

# add a legend
ax.legend(loc='best', frameon=False)

# save the plot
plt.savefig('Talk about an inefficient estimator.png')

# display plot
plt.show()
```

The extreme heteroskedasticity implies the OLS estimator is inefficient.  This inefficiency will impact the distribution of our $p$-values as we shall see...

```python
# p-values associated with the test statistics from our MC
p_values = 2 * stats.distributions.t.sf(t_stats.abs(), N - 2)
```

```python
# create a new figure
fig=plt.figure()
ax = fig.add_subplot(111)

# generate the histogram
n, bins, patches = ax.hist(p_values, color='yellow', bins=40, normed=True, alpha=0.75)
ax.set_xlabel('p-Values')
ax.set_ylabel('Density')
ax.set_title('Bad size properties!')

# save the figure
plt.savefig('Bad size properties.png')

# display 
plt.show()
```

Note that the above histogram places too much weight on low $p$-values.  This means that we will be over rejecting the true null hypothesis (i.e., too much type I error).

```python
# critica value for 5% test
CV = stats.distributions.t.ppf(1 - (0.05 / 2), N - 2)

# how many times did we reject the null (remember good size properties would yield something close to 250)
t_stats[ t_stats > CV ].count()
```
