Replicability is an essential part of empirical science. If experimental results can’t be replicated by other researchers, then as far as the rest of the research community is concerned, they might as well not have happened in the first place. The same principle holds in empirical work in economics. At the very minimum, for a given dataset, you must be able to replicate your original results.
This is good practice for another reason: you always can’t realistically expect to get the answer to an empirical research problem in a single computer session. How do we interpret the entries for the lngdp60 row? The estimated coefficient is .0037724. What does this mean? The dependent variable is growth, which is the average annual log growth rate (not the percentage growth rate). The explanatory variable is lngdp60, which is the log of the value of GDP per worker at the start of the period, 1960. To fix magnitudes, here is a summary of these two variables:

```python
# summarize growth
mrw_1992_data_nonoil.growth.describe()
```

So a mean of 0.018 for growth means that the average annual growth rate for this period was 1.8% per annum.

```python
# summarize log GDP per capita in 1960
mrw_1992_data_nonoil.lngdp60.describe()
```

Now consider the estimated coefficient. The value is approximately 0.0038. If lngdp60 increased by 1 unit, then growth would increase by 0.0038. Consider these magnitudes in turn. A 1-unit increase in lngdp60 is a very big increase in GDP per worker. To be precise, since $e^1= e \approx 2.72$ (we are raising e to the power 1 because we are considering a 1-unit increase in a natural log), it corresponds to a 100 * (2.72 - 1) = 172% increase in GDP per worker, i.e., between doubling (100% increase) and tripling (200% increase). Meanwhile, the change in the growth rate is 1 * 0.0038 = 0.0038 = 0.38%. We can say that “if GDP per worker in 1960 goes up by 172%, the annual growth rate goes up by 0.38 percentage points”.

Note: if you are working with a RHS variable that is a log, you might find it convenient to think in terms of a doubling in the underlying variable. Since ln(2) = 0.69, and 0.69 * 0.0038 = 0.0026, we can say “if GDP per worker in 1960 doubles, the annual growth rate goes up by 0.26 percentage points”.

So much for interpretation. How precise is the estimate? This is what the rest of the line tells us about. The easiest way to think about the precision of the coefficient estimate is in terms of the confidence interval. 

```python
# get 95% confidence intervals for each regressor 
results.conf_int(alpha=0.05)
```

The last two figures in the row are the endpoints of a 95% confidence interval: [-0.000167 , 0.007712]. Our model is

$$growth = \alpha +\beta * lngdp60$$

$\beta$ is a population parameter (i.e., we don’t know what it's true value is). Intuitively, and very roughly speaking, we are “95% confident” that the true value of $\beta$ lies in the interval generated by our OLS estimation: [-0.0001671 , 0.007712].

This is a rather wide confidence interval. Say we were still interested in the impact of a 1-unit increase in lngdp60. We are “95% confident” that the impact of this is somewhere between a *fall* in the annual growth rate of -0.017 percentage points, and an increase in the annual growth rate of 0.78 percentage points.

The columns for the standard error, t-statistic and p-value have essentially the same information about the precision of the estimated parameter as the confidence interval, just presented differently. The null hypothesis being tested is:

$$H_{0}: \beta=0$$

The OLS estimate of the parameter, $\hat{\beta}$, is a statistic, and as such it has a variance. The square root of the estimated variance of $\hat{\beta}$ is called its "standard error."  Under the null hypothesis, the ratio

$$\frac{\hat{\beta}}{SE(\hat{\beta})}$$

has a t-distribution with N-K degrees of freedom, where K is the number of regressors including the constant (here, K=2). This ratio is our “test statistic”. The entry in the “t” column in the regressor table above is simply this ratio. 

```python
# get the t-statistics for each regressor
results.tvalues
```

Finally, if the null hypothesis is correct (i.e., the true value of $\beta$ really is zero) then we know this test statistic comes from a t-distribution. This means we can calculate the probability that we would observe a test statistic with at least this value, or the “p-value.” By convention, we normally calculate this for alternatives in which the true $\beta$ is either very positive or very negative, i.e., we usually work with 2-tailed tests. By default, statsmodel reports the p-value using a 5% significance level.

```python
# get the p-values for the standard significance test of each regressor
results.pvalues
```

A simple and intuitive way to link the t-test and the confidence interval is to consider whether 0 appears in the confidence interval. If 0 is **not** in the 95% confidence interval for a given regressor, then the reported p-value will be *less* than 5%. If 0 is in the 95% confidence interval, then the reported p-value will be *greater* than 5%.

The t-distribution is almost indistinguishable from the normal as soon as the degrees of freedom are more than 30 or so, so in practice you will usually get virtually the same results if you assume the ratio

$$\frac{\hat{\beta}}{SE(\hat{\beta})}$$

has a normal distribution. And since the critical value for a 2-tailed test using the Normal is 1.96, you can usually work out an approximate confidence interval by adding/subtracting $2*SE(\hat{\beta})$ to the estimated coefficient. Similarly, you can usually conduct an approximate test of the null hypothesis that $\beta = 0$ by seeing whether the above ratiois greater than 2 or not.

How do we interpret the constant? One way to think of it is that it is an estimate of the conditional mean of the dependent variable, in this case $growth$. But it is an estimate that is usually not calibrated in a way that is particularly useful. The estimated constant answers the question: “Say we know the log level of GDP per worker in 1960. What is our best estimate of the growth rate if the log level of GDP per worker were zero?”

This is an answer to a not-particularly-useful question. On the other hand, the answer to the following question is even less useful, because we know in advance what the answer is: “Say we know the log level of GDP per worker in 1960. What is our best estimate of the growth rate if the log level of GDP per worker were the average for the sample?”

```python
# create a new variable representing the de-meaned lngdp60
mrw_1992_data_nonoil['lngdp60_c'] = mrw_1992_data_nonoil['lngdp60'] - mrw_1992_data_nonoil.lngdp60.mean()
```

###Basic regression statistics

First, the $R^2$.  The exact value of the $R^2$ for a regression can be retrived as follows...  

```python
# get the r-squared of the regression
results.rsquared
```

The $R^2$ of a regression is a measure of "Goodness-of-fit" (i.e., how well the model fits the data).  To see where the $R^2$ comes from (and why it can be thought of as a measure of goodness-of-fit) start with our basic model:

$$growth_{i} = \alpha + \beta * lngdp60 + \epsilon_{i}$$

where $\epsilon_{i}$ is the error term (i.e., everything we aren't modelling). The estimated model with fitted values of the dependent variable:

$$\widehat{growth} = \hat{\alpha} + \hat{\beta} * lngdp60$$

```python
# get the first few fitted values 
results.fittedvalues.head()
```

The residuals are the differences between the actual dependent variable and what the model predicts for it based on the actual values of the regressors

$$\hat{\epsilon} = growth_{i} - \widehat{growth}_{i}$$

```python
# get the first few residuals
results.resid.head()
```

The definition of $R^2$ is the share of total variation in the dependent variable explained by the model. It can be calculated using either the fitted values of the dependent variable or using the residuals:

$$ R^2 = \frac{ESS}{TSS} = 1 - \frac{RSS}{TSS} $$

```python
# compare the hand calculation for R-squared
results.ess / results.centered_tss
```

```python
# with the actual R-squared
results.rsquared
```

Final comment, the $adjusted\ R^2$ reported in the regression table above corrects the traditional $R^2$ to account for the number of independent variables.

The $F\ Statistic$ and $Prob\ (F\ Statistic)$ report the test statistic and $p$-value associated with the standard F-Test.  The F-test is a test of the joint significance of all K regressors **not** counting the constant. The reason we exclude the constant is by convention; the F-test is a test of the “model”, and we don’t consider whether or not the dependent variable has a non-zero mean as an important component of our model. In this case, K = 2, so the F test has K - 1=1 numerator degree of freedom. The denominator degrees of freedom = N - K = 98 - 2 = 96.

```python
# dfs for the model (i.e., numerator of the F-test)
results.df_model
```

```python
# dfs for the model (i.e., denominator of the F-test)
results.df_resid
```

```python
# test statistic for the standard F test
results.fvalue
```

```python
# p-value for the standard F-test
results.f_pvalue
```

Although not reported in the regression table above, one can easily calculate the root mean squared error of the regression.  Recall that the RMSE is basically the average residual.  Except that instead ov dividing by $N$, we divide correct for the degrees of freedom by dividing by $N-K$. Thus the formula for the RMSE is

$$ RMSE = \sqrt{\frac{\sum \hat{\epsilon}^2_{i}}{N - K}} $$

```python
# computes the RMSE using the residual sum of squares
np.sqrt((1 / 96.) * results.ssr)
```

The RMSE is an unbiased and consistent estimate of $\sigma$, the square root of the variance of the error $\epsilon_{i}$.  Question: is there any difference asymptotically between the traditional formula for estimating $\sigma$ above (i.e., RMSE), and the following simplified formula? 

$$ \hat{\sigma}= \sqrt{\frac{\sum \hat{\epsilon}^2_{i}}{N}} $$

Might there be a difference in "small" samples?  Recall that the t-distribution $N-K$ degrees of freedom is well approximated by the Normal distribution so long as $N - K \ge 30$. Same idea is being applied here, so long as our sample is not too small, we can get away with estimating the RMSE or $\hat{\sigma}$ using the simpler formula.

How do you interpret the sign and size of the coefficient on the main regressor of interest, $lngdp60$? consider the estimated coefficient...

```python
results3.params['lngdp60']
```

The sign is now negative (in agreement with theory!).  If $lngdp60$ *decreased* by 1 unit, then $growth$ would *increase* by $\approx$ 0.011. Consider these magnitudes in turn. Recall that a one unit decrease in $lngdp60$ is a very big increase in GDP per worker. To be precise, since $e^1 = e \approx 2.72$ (we are raising $e$ to the power 1 because we are considering a 1-unit decrease in a natural log), it corresponds to a 100 * (2.72 - 1) = 172% *decrease* in GDP per worker! The change in the growth rate is 1 * -0.011 = -0.011 = -1.1%. We can say that “if GDP per worker in 1960 goes down by 172%, the annual growth rate goes up by 1.1 percentage points”. 

More convenient to think in terms of a halving of the underlying variable. Since ln(0.5) = -0.69, and -0.69 * -0.011 = 0.0076, we can say “if GDP per worker in 1960 was cut in half, the annual growth rate would up by 0.76 percentage points”. Poorer countries would grow faster (i.e., conditional convergence).
