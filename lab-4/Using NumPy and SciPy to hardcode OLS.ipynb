```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
import pandas as pd

# remember to set the max_columns to get data to display properly!
pd.set_printoptions(max_columns=99)

# set the print options for numpy
np.set_printoptions(linewidth=120)
```

```python
# import the Mankie, Romer, and Weil (1992) data
mrw_1992_data = sm.iolib.genfromdta('mrw1992.dta', missing_flt=NaN, missing_str=NaN, pandas=True)
```

```python
# display the data
mrw_1992_data
```

To make it easier to play around with the data, we will work with a subset of the dataset: 22 OECD countries only.  Subset the data accordingly...

```python
# drop all non-oecd obs
mrw_1992_data_OECD = mrw_1992_data[mrw_1992_data['oecd'] == 1]

# display the data
mrw_1992_data_OECD
```

Create variables representing the natural logarithm of $gdp60$ and $gdp85$, and a third variable representing the average annual growth rate of per capita real GDP... 

```python
# create some new variables using the natural logarithm function from NumPy
mrw_1992_data_OECD['lngdp60'] = np.log(mrw_1992_data_OECD['gdp60'])
mrw_1992_data_OECD['lngdp85'] = np.log(mrw_1992_data_OECD['gdp85'])
mrw_1992_data_OECD['growth'] = (mrw_1992_data_OECD['lngdp85'] - mrw_1992_data_OECD['lngdp60']) / 25.
```

Run a couple of regressions.  First, a simply bivariate regression of $growth$ on a constant and $lngdp60$...

```python
# our endogenous (or dependent or response variable) is growth
y = mrw_1992_data_OECD['growth']

# our matrix of exogenous (or independent variables) are a constant and lngdp60
X = sm.add_constant(mrw_1992_data_OECD['lngdp60'], prepend=True)

# run the regression
results = sm.OLS(endog=y, exog=X).fit()

# examine the output
print results.summary()
```

The coefficient on $lngdp60$ is negative and significant: if you were poor in 1960, then you grew faster (on average) over the next 25 years which suggest unconditional convergence. However, in additional to being small, we have biased our sample significantly by including only OECD countries. 

Create a scatter plot of the simple bivariate regression (just to remind yourself how!)...

```python
fig = plt.figure()
ax = fig.add_subplot(111)

X = mrw_1992_data_OECD.lngdp60.values
Y = mrw_1992_data_OECD.growth.values

# the scatterplot
ax.scatter(X, Y, color='orange', marker='o')

# doesn't seem to be a command to plot a line using slope and intercept!
grid = np.linspace(5, 10.5, 1000)
ax.plot(grid, results.params[1] * grid + results.params[0], color='red', label='OLS')

ax.set_xlim(5.5, 10.5)
ax.set_xlabel('log real GDP per worker (1960)')
ax.set_ylabel('Average growth in GDP 1960-1985')
ax.set_title('Sample bias anyone?')

ax.legend(loc='best', frameon=False)

plt.show()
```

...then, run the multiple regression including $popgrowth$, $igdp$, and $school$ as regressors (along with $lngdp60$)...

```python
# our endogenous (or dependent or response variable) is growth
y = mrw_1992_data_OECD['growth']

# our matrix of exogenous (or independent variables) are a constant and lngdp60
X = sm.add_constant(mrw_1992_data_OECD[['lngdp60', 'popgrowth', 'igdp', 'school']], prepend=True)

# run the regression
results2 = sm.OLS(endog=y, exog=X).fit()

# examine the output
print results2.summary()
```

Interpretation of the above is similar to the simple bivariate regression.  Note that the coefficient on $lngdp60$ is now larger in magnitude.

Task 5: The Frisch-Waugh-Lovell Theorem
---------------------------------------

The [Frisch-Waugh-Lovell Theorem](http://en.wikipedia.org/wiki/Frisch%E2%80%93Waugh%E2%80%93Lovell_theorem) is an important and useful theorem with a wide range of applications. Other terminology used for the FWL Theorem is “partialling out” and “regression anatomy”.

The theorem states the following. Say we have a model we wish to estimate using OLS. The dependent variable $y$ and the disturbance term $\epsilon$ are both $N \times 1$, and the matrix of regressors $X$ is $N \times K$.

$$ y = X\beta + \epsilon$$

And if we were to use OLS to estimate $\beta$, we would obtain

$$\hat{\beta} = (X'X)^{-1}X'y $$

Partition the matrix of regressors $X$ into the $K_1$ regressors that we are not interested in and the $K_2$ regressors of interest. (Say $X_1$ are controls that we think we need to include in the model but in which we have no interest: seasonal, industry or geographical dummies, demographic controls, etc.) Partition the set of coefficients conformably. We have:

$$y = X_1\beta_1 + X_2\beta_2 + \epsilon$$

We want to “partial out” the uninteresting regressors $X_1$. The FWL Theorem says we can do this and obtain the interesting coefficients $\beta_2$ without having to estimate the uninteresting coefficients $\beta_1$ as well. It works as follows. Define the projection and annihilation matrices for the uninteresting regressors $X_1$:

$$P_{1} = X_1(X_1'X_1)^{-1}X_1' $$
$$M_{1} = I - P_{1} $$

Now define

1. $$\tilde{y} = M_{1}y $$
2. $$\tilde{X_2} = M_{1}X_{2}$$

Note that I is simply the residuals from a regression $y$ on the uninteresting regressors $X_1$, and II is the residuals from a regression of $X_2$ on our uninteresting regressors $X_1$!

The FWL Theorem states that the OLS coefficient on the $K_2$ interesting regressors $X_2$ obtained by estimating the full regression $y = X\beta + \epsilon$ is exactly the same as the OLS coefficient obtained by estimating

$$\tilde{y} = \tilde{X_2}\beta_2 + u$$

and obtaining

$$\hat{\beta_2} = (\tilde{X_2}'\tilde{X_2})^{-1}\tilde{X_2}'\tilde{y}$$

Moreover, the variance-covariance matrix obtained by this OLS estimation will also be the same as that obtained by OLS estimation of the full equation, except that, if the OLS estimation includes a finite sample adjustment (statsmodels does!), the degrees of freedom will be wrong: instead of using the standard $N-K$, it will use $N-K_2$.

In this task, you will illustrate the FWL theorem by estimating the full MRW model but only in order to obtain just the coefficient on log GDP per capita in 1960. The other regressors – population, I/Y ratio, schooling, and the constant.

First we need to create the appropriate matrices...

```python
# just renaming the our matrices from above to conform with FWL
y, X1, X2 = growth, np.hstack(tup=(ones, pop, igdp, school)), lngdp60
```

Next, create the annihilation matrix $M_1$. Make sure it’s symmetric...

```python
# first create the projection matrix...
P1 = X1.dot(np.linalg.inv((X1.T).dot(X1))).dot(X1.T)

# then create thee annihilation matrix
M1 = np.eye(22) - P1

# then make sure its symmetric
M1 = 0.5 * (M1 + M1.T)
```

Create the new dependent variable $\tilde{y}$ and matrix of regressors $\tilde{X_2}$ from above...

```python
# we are partialling out X1!
y_tilde = M1.dot(growth)
X2_tilde = M1.dot(X2)
```

And, finally, create and display the OLS coefficient on $X_2$...

```python
# OLS estimator for beta_2
beta_2 = np.linalg.inv((X2_tilde.T).dot(X2_tilde)).dot(X2_tilde.T).dot(y_tilde)

# display the result...
print beta_2
```

```python
# compare with regression results from above...
results2.params['lngdp60']
```

The FWL procedure suggests we can learn from a bivariate scatterplot of $\tilde{y}$ on $\tilde{X_2}$ in the same way that we learned from a scatterplot of $y$ on $X_2$ when $X_2$ (i.e., log GDP per capita in 1960) was the only explanatory variable in the model. This intuition is correct!  

First, run a simple bivariate regression of growth on lngdp60 for all countries in the sample (Note you will need to recreate some variables!)

```python
# create lngdp variables
mrw_1992_data['lngdp60'] = np.log(mrw_1992_data['gdp60'])
mrw_1992_data['lngdp85'] = np.log(mrw_1992_data['gdp85'])

# create the growth variable
mrw_1992_data['growth'] = (mrw_1992_data['lngdp85'] - mrw_1992_data['lngdp60']) / 25.
```

```python
# alternative regression specification using pd.ols which has built-in missing value handling.  
# looks like statsmodels will have this capability as well in version 0.5 and long term plan is too merge Pandas and Statsmodels
model = pd.ols(y=mrw_1992_data['growth'], x=mrw_1992_data[['lngdp60']])

print model
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

X = mrw_1992_data.lngdp60.values
Y = mrw_1992_data.growth.values

# the scatterplot
ax.scatter(X, Y, color='orange', marker='o')

# doesn't seem to be a command to plot a line using slope and intercept!
grid = np.linspace(5, 12, 1000)
ax.plot(grid, model.beta['lngdp60'] * grid + model.beta['intercept'], color='red', label='OLS')

ax.set_xlim(5,12)
ax.set_xlabel('log real GDP per worker (1960)')
ax.set_ylabel('Average growth in GDP 1960-1985')
ax.set_title('Basically no relationship...')

ax.legend(loc='best', frameon=False)

plt.show()
```

Applying the FWL theorem we can partial out the uninteresting regressors as follows...

```python
# regress y on uninteresting redressors and store the residuals as y_tilde
y_tilde = pd.ols(y=mrw_1992_data['growth'], x=mrw_1992_data[['popgrowth', 'igdp', 'school']]).resid

# regress lngdp60 on uninteresting regressors and store as residuals X2_tilde
X2_tilde = pd.ols(y=mrw_1992_data['lngdp60'], x=mrw_1992_data[['popgrowth', 'igdp', 'school']]).resid
```

Create a scatter plot of $\tilde{y}$ and $\tilde{X_2}$ (this is equivalent to STATA's *avplot* command)...

```python
fig = plt.figure()
ax = fig.add_subplot(111)

X = X2_tilde.values
Y = y_tilde.values

# the scatterplot
ax.scatter(X, Y, color='orange', marker='o')

# regress y_tilde on X2_tilde
FWL = pd.ols(y=y_tilde, x=X2_tilde)

# doesn't seem to be a command to plot a line using slope and intercept!
grid = np.linspace(-2, 3, 1000)
ax.plot(grid, FWL.beta['x'] * grid + FWL.beta['intercept'], color='red', label='OLS')

ax.set_xlim(-2,3)
ax.set_xlabel(r'E[lngdp60 | $X_1$]')
ax.set_ylabel(r'E[growth | $X_1$]')
ax.set_title('Conditional convergence via the FWL Theorem!')

ax.legend(loc='best', frameon=False)

plt.show()
```
